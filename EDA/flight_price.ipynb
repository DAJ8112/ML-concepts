```python
def convertToMinutes(duration):
    parts = duration.split(' ')
    hour = int(parts[0].replace('h', '')) if 'h' in parts[0] else 0
    minute = int(parts[1].replace('m', '')) if len(parts) > 1 else 0
    return hour * 60 + minute
```

```python
df['duration'] = df['Duration'].apply(convertToMinutes)
```
